```python
import requests
```

```python
import bs4
```

```python
request = requests.get("https://www.flipkart.com/poco-m4-pro-5g-cool-blue-64-gb/p/itm827b338897bfe")
```

```python
request
```

```python
soup = bs4.BeautifulSoup(request.text)
```

```python
soup
```

```python
reviews = soup.find_all('div',{'class':'t-ZTKy'});
for review in reviews:
    print(review.get_text()+ "\n\n")
```

```python
ratings = soup.find_all('div',{'class':'_3LWZlK _1BLPMq'});
for rating in ratings:
    print(rating.get_text()+"\n\n")
```

```python
rating1 = soup.find('div',{'class':'_2d4LTz'}).get_text()
print(rating1)
```

```python
cust_name = soup.find_all('p',{'class':'_2sc7ZR _2V5EHH'});
for customer in cust_name:
    print(customer.get_text()+"\n\n")
```
